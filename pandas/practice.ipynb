#### vectorization

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({"Number": [1, 2, 3, 4, 5]})

df['Type'] = np.where(df['Number'] % 2 == 0, 'Even', 'Odd')

print(df)
```

#### np.select

```python
df = pd.DataFrame({"Age": [12, 25, 45, 70]})

conditions = [
    df["Age"] < 18,
    df["Age"].between(18, 60),
    df["Age"] > 60
]

choices = ["Minor", "Adult", "Senior"]

df["Category"] = np.select(conditions, choices, default="Unknown")
print(df)   
```

#### Kadane's Algorithm

```python
"""def sub_array(num_list):
    length = len(num_list)
    sum_list = []
    for i in range(length):
        for j in range(i+1, length+1):
            print(num_list[i:j])
            sum_list.append(sum(num_list[i:j]))
    return max(sum_list)

sub_array([1,2,3])"""

# Kadane's Algorithm

def max_subarray(nums):
    max_sum = nums[0]
    current_sum = nums[0]

    for i in range(1, len(nums)):
        current_sum = max(nums[i], current_sum + nums[i])
        max_sum = max(max_sum, current_sum)

    return max_sum

print(max_subarray([-2,1,-3,4,-1,2,1,-5,4]))
```

#### Two pointer strategy

```python
def pair_with_sum(arr, target):
    left = 0
    right = len(arr)-1

    while left < right:
        current_sum = arr[left] + arr[right]

        if current_sum == target:
            return True, (arr[left], arr[right])
        elif current_sum < target:
            left += 1
        else:
            right -= 1
    
    return False, None

arr = [1,2,3,4,5,6,7]
target = 5

found, pair = pair_with_sum(arr, target)

if found:
    print("Pair found:", pair)
else:
    print("No pair found")
```

#### find the **first repeating element**

```python
# (index-based first repeat)

arr = [10, 5, 3, 4, 3, 5, 6]

found = False

for i in range(len(arr)):
    for j in range(i + 1, len(arr)):
        if arr[i] == arr[j]:
            print("First repeating element is:", arr[i])
            found = True
            break
    if found:
        break
```

```python
seen = set()

for x in arr:
    if x in seen:
        print("First repeating element is:", x)
        break
    seen.add(x)
```

```python
arr = [10, 5, 3, 4, 3, 5, 6]

index_map = {}
min_index = len(arr)
result = None

for i, x in enumerate(arr):
    if x in index_map:
        if index_map[x] < min_index:
            min_index = index_map[x]
            result = x
    else:
        index_map[x] = i

print("First repeating element is:", result)
```

```python
arr = [10, 5, 3, 4, 3, 5, 6, 10, 4, 6]

freq = {}

# Count frequency
for x in arr:
    freq[x] = freq.get(x, 0) + 1

# Find first non-repeating
for x in arr:
    if freq[x] == 1:
        print("First non-repeating element:", x)
        break
else:
    print("All elements are repeating")
```

#### stack balancing the brackets

```python
def is_balanced(s):
    stack = []
    pairs = {')': '(', '}': '{', ']': '['}

    for ch in s:
        if ch in "({[":
            stack.append(ch)
        elif ch in ")}]":
            if not stack or stack[-1] != pairs[ch]:
                return False
            stack.pop()

    return len(stack) == 0


print(is_balanced("{[()]}"))  # True
```
