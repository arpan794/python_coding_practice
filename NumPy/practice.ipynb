### Indexing and Slicing

```python
arr = np.random.randint(1,10,(3,5))
print(arr)
```

```python
arr[0]
arr[1:2]
arr[:,-1]
arr[0,1]
```

### Reshaping Arrays

```python
arr = np.random.randint(0,100,(3,5))
print(arr)
```

```python
arr.reshape(5,3)
```

```python
arr.flatten()
```

```python
arr.ravel() # ravel is similar to flatten but it returns a view of the original array if possible, while flatten always returns a copy.
```

### Mathematical Operations

```python
arr = np.random.randint(0,100,(3,5))
print(arr)
```

```python
np.sum(arr)
np.mean(arr)
np.median(arr)
np.std(arr)
np.min(arr)
np.max(arr)
np.argmin(arr) # return the index of minimum value
np.argmax(arr) # return the index of maximum value
np.argsort(arr) # return the indices that would sort the array
```

```python
print(arr)
```

```python
np.sum(arr, axis=0) # sum along columns
```

```python
np.sum(arr, axis=1) # sum along rows
```

### Broadcasting

```python
arr = np.random.randint(0,100,(3,5))
print(arr)
```

```python
arr + 5
```

### Boolean Masking

```python
arr_mas = arr[arr>50] # boolean indexing
print(arr_mas)
```

```python
arr[ arr % 2 == 0 ] # boolean indexing for even numbers
```

### Random Modules

```python
np.random.seed(42) # is used to control randomness so that your results are reproducible.
arr1 = np.random.randn(3,3)
print(arr1)
arr2 = np.random.rand(3,3)
print(arr2)
arr3 = np.random.randint(1,10,5)
print(arr3)
```
